```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

plt.style.use('seaborn-v0_8')

df = pd.read_csv('/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/laptop_price.csv')
df.head(2)
```

```python
df['Price (Euro)'].plot.hist(bins=40, title='Distribusi Harga Laptop (Euro)')

plt.xlabel('Harga (Euro)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
df.plot(kind='scatter', x='Weight (kg)', y='Price (Euro)', title='Harga Laptop Berdasarkan Beratnya')

plt.show()
```

```python
df.groupby('Company')['Price (Euro)'].mean().plot(kind='barh', title='Rata-rata Harga Laptop Setiap Perusahaan', color = 'gold')
plt.show()
```

```python
fig, axs = plt.subplots(1, 6, figsize=(20, 5))

for i, tipe in enumerate(df['TypeName'].unique()):
    df[df['TypeName'] == tipe].plot(kind='box', y='Price (Euro)', ax=axs[i], title=f'Boxplot Harga \nLaptop ({tipe})')
    axs[i].set_ylim(0,7000)

plt.show()
```
