```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

plt.style.use('ggplot')

df = pd.read_csv('/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/dummy_ojol_transactions.csv', sep=';')
df.head(4)
```

```python
df.plot(kind = 'box', y = 'amount_delivery')
```

```python
df.plot(kind = 'box', y = 'amount_merchant')
```

```python
# filtering untuk tipe food dan shop
df2 = df[(df['mode'] == 'FOOD') | (df['mode'] == 'SHOP')]

# filtering untuk tipe bike dan car
df3 = df[df['mode'].isin(['BIKE', 'CAR'])]

# Membuat boxplot
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# boxplot jumlah merchant
df2['amount_merchant'].plot.box(color='green', ax=ax1)

ax1.set_title('Box Plot Jumlah Merchant')
# plt.ylabel('Jumlah Pengiriman')
# plt.show()

# boxplot jumlah delivery
df2['amount_delivery'].plot.box(color='black', ax=ax2)

ax2.set_title('Box Plot Jumlah Delivery')
# plt.ylabel('Jumlah Pengiriman')
# plt.show()

# boxplot jumlah pengantaran
df3['amount_delivery'].plot.box(color='purple', ax=ax3)

ax3.set_title('Box Plot Jumlah Pengantaran')
# plt.ylabel('Jumlah Pengiriman')
# plt.show()
```
