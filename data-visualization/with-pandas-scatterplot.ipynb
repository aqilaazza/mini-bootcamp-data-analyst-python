```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

plt.style.use('ggplot')

df = pd.read_csv('/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/dummy_ojol_transactions.csv', sep=';')
df.head(4)
```

```python
df.info()
```

```python
df['distance'] = df['distance'].str.replace(',', '.').astype(float)
df.info()
```

```python
df
```

```python
fig, ax = plt.subplots()

df[df['mode'] == 'BIKE'].plot(kind='scatter', x='distance', y='transaction_amount_total', color='g', label='BIKE', ax=ax)
df[df['mode'] == 'CAR'].plot(kind='scatter', x='distance', y='transaction_amount_total', color='r', label='CAR', ax=ax)

fig.show()
```

Semakin banyak distancenya maka akan semakin banyak total transaksinya
