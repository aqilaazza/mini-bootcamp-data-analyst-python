```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

plt.style.use('ggplot')

df = pd.read_csv('/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/dummy_ojol_transactions.csv', sep=';')
df.head(4)
```

```python
df.info()
```

```python
df['date_order'] = pd.to_datetime(df['date_order'], format='%d/%m/%Y %H.%M', errors='coerce')
```

```python
df['date'] = df['date_order'].dt.date
```

```python
df.head()
```

```python
df.groupby(by='date')['transaction_amount_total'].sum().plot(kind='line')
```

```python
df.groupby(by='date')['transaction_amount_total'].mean().plot(kind='line')
```
