# Visualisasi Data dengan Pandas dan Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

plt.style.use('ggplot')

df = pd.read_csv('/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/dummy_ojol_transactions.csv', sep=';')
df.head(4)
```

#Histogram

Cara 1

```python
df.plot.hist(column='transaction_amount_total', bins = 20)

plt.title("Histogram Total Transaksi OJOl")
plt.xlabel("Total Tarnsaksi")

plt.show()
```

Cara 2

```python
df['transaction_amount_total'].plot.hist(bins=20, color='lime')

plt.title("Histogram Total Transaksi OJOl")
plt.xlabel("Total Tarnsaksi")

plt.show()
```

Cara 3

```python
df['transaction_amount_total'].plot(kind='hist', bins=20, color='lightblue')

plt.title("Histogram Total Transaksi OJOl")
plt.xlabel("Total Tarnsaksi")

plt.show()
```

#Visualisasi Data dengan Pandas Berbasis Objek

```python
plot_total = df['transaction_amount_total'].plot(kind='hist', bins=20, color='lightblue', figsize=(10,5))
type(plot_total)
```

```python
plot_total = df['transaction_amount_total'].plot(kind='hist', bins=20, color='lightblue')

plot_total.set_title('Histogram Total Transaksi OJOL')
plot_total.set_xlabel('Total Transaksi')

plt.show()
```
