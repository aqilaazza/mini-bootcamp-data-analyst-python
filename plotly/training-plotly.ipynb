```python
import plotly.express as px

tips = px.data.tips()
tips.head()
```

1. Buatlah scatter plot seperti di bawah ini!

```python
fig = px.scatter(tips,
                 x="tip",
                 y="total_bill",
                 size='size')
fig.show()
```

2. Buatlah scatter plot seperti di bawah ini!

```python
fig = px.scatter(tips,
                 x="tip",
                 y="total_bill",
                 color="sex",
                 facet_col="smoker",
                 trendline='ols')
fig.show()
```

3. Buatlah scatter plot seperti di bawah ini!

```python
fig = px.scatter(tips,
                 x="size",
                 y="total_bill",
                 color="day",
                 facet_col="day",
                 size = 'size')
fig.show()
```

4. Buatlah scatter plot seperti di bawah ini!

```python
fig = px.scatter(tips,
                 x="tip",
                 y="total_bill",
                 color="time",
                 facet_col="smoker",
                 facet_row="sex",
                 size = 'size')
fig.show()
```
