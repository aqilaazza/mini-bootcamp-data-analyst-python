#Dasar-dasar Plot Plotly

Plotly menyediakan beberapa modul yang bisa digunakan untuk membuat visualisasi data, diantaranya adalah plotly express dan plotly graph object. Plotly Express adalah modul dalam Plotly yang dirancang untuk memudahkan pembuatan berbagai visualisasi interaktif dan menarik dengan sedikit kode (simple dan mudah), sedangakan dengan plotly graph object memberikan keleluasaan untuk melalukan modifikasi. Bagi pemula yang ingin belajar membuat visualisasi dari plot disarankan untuk menggunakan plotly express karena mudah dimengerti.

```python
import plotly.express as px
```

Fungsi Plotly Express mengikuti sintaks yang konsisten:

px.function_name(data_frame, x='x_column', y='y_column', ...).

Parameter data_frame adalah dataset, dan x serta y menentukan kolom untuk sumbu x dan y.

#Dataset Plotly

Plotly menyediakan beberapa dataset bawaan yang dapat digunakan untuk eksplorasi dan visualisasi data. Dataset-dataset ini terintegrasi dengan baik dalam library Plotly, terutama dengan modul Plotly Express.

```python
# Menggunakan dataset iris dari Plotly
df = px.data.iris()
df.head()
```

#Simple Plot dengan Dataframe

```python
px.scatter?
```

```python
px.scatter(df, x="sepal_width", y="sepal_length")
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

# Menggunakan Parameter Color

```python
# Menambahkan parameter color categorical
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
# Menambahkan parameter color continue
fig = px.scatter(df, x="sepal_width", y="sepal_length", color='petal_length')
fig.show()
```

```python
# Mengubah jenis color scale
fig = px.scatter(df, x="sepal_width", y="sepal_length", color='petal_length', color_continuous_scale='Viridis')
fig.show()
```

#Menambahkan Parameter Symbol

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", symbol="species")
fig.show()
```

#Menggunakan Parameter Size

```python
# Menambahkan parameter color categorical
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", size='petal_length')
fig.show()
```

#Memodifikasi Hover Data

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", hover_data=['petal_length', 'petal_width'], hover_name='species')
fig.show()
```

#Scatter Plot dengan Data Kategorikal

```python
df2 = px.data.medals_long()
df2.sample(3)
```

```python
fig = px.scatter(df2, y="nation", x="count", color="medal", symbol="medal")
fig.show()
```

#Menggunakan error bar

```python
df["e"] = df["sepal_width"]/100
fig = px.scatter(df,
                 x="sepal_width",
                 y="sepal_length",
                 color="species",
                 error_x="e",
                 error_y="e")
fig.show()
```

#Menggunakan Marginal Distribution

```python
fig = px.scatter(df,
                 x="sepal_width",
                 y="sepal_length",
                 marginal_x="histogram",
                 marginal_y="rug")
fig.show()
```

```python
fig = px.scatter(df,
                 x="sepal_width",
                 y="sepal_length",
                 color='species',
                 marginal_x="histogram",
                 marginal_y="rug")
fig.show()
```

#Facetting

```python
fig = px.scatter(df,
                 x="sepal_length",
                 y="sepal_width",
                 color="species",
                 facet_col="species")
fig.show()
```

```python
fig = px.scatter(df,
                 x="sepal_length",
                 y="sepal_width",
                 color="species",
                 facet_row="species")
fig.show()
```

#Menggunakan Trendline

```python
fig = px.scatter(df,
                 x="sepal_length",
                 y="sepal_width",
                 color="species",
                 facet_row="species",
                 trendline='ols')
fig.show()
```
