```python
import plotly.express as px

gapminder = px.data.gapminder()
```

#Line Plot

```python
# Data dari negara asean
negara_asean = ['Indonesia', 'Malaysia', 'Singapore', 'Thailand', 'Vietnam', 'Philippines', 'Myanmar', 'Laos', 'Cambodia', 'Brunei', 'Timor-Leste']
asean = gapminder[gapminder['country'].isin(negara_asean)]
```

```python
fig = px.line(asean,
              x='year',
              y='lifeExp',
              color='country',
              title='Contoh Line Plot')
fig.show()
```

#Bar Plot

```python
# Mengambil data indonesia
indonesia = gapminder[gapminder['country'] == 'Indonesia']
```

```python
fig = px.bar(indonesia, x='year', y='gdpPercap', title='Contoh Bar Chart')
fig.show()
```

#Bar Stacked Chart

```python
fig = px.bar(asean, x='year', y='gdpPercap', color='country', title='Contoh Bar Chart')
fig.show()
```

```python
# Dengan menggunakan facet
fig = px.bar(asean, x='year', y='gdpPercap', color='country', facet_col='country', title='Contoh Bar Chart')

fig.show()
```

#Histogram

```python
fig = px.histogram(gapminder, x='gdpPercap', title='Contoh Histogram', nbins=50)
fig.show()
```

#Boxplot

```python
fig = px.box(gapminder, x='continent', y='lifeExp', color='continent', title='Contoh Boxplot')
fig.show()
```

#Plot-plot Lainnya

Plot lainnya dapat dilihat di: https://plotly.com/python-api-reference/plotly.express.html#px

```python
asean.head()
```

```python
px.scatter_3d(asean, x='year', y='gdpPercap', z='lifeExp', color='country', size='pop')
```

```python
px.choropleth(gapminder, locations='iso_alpha', color='lifeExp',animation_frame='year' )
```

```python
# create a map using choropleth
fig = px.choropleth(gapminder.query('year == 2007'), locations='iso_alpha', color='lifeExp', hover_name='country',
                    projection='natural earth')
fig.show()
```

```python
fig = px.violin(gapminder, x='continent', y='lifeExp', color='continent', points='all', title='Contoh Violin Plot')
fig.show()
```

```python
fig = px.scatter_3d(asean,
                    x='year',
                    y='lifeExp',
                    z='gdpPercap',
                    color='country',
                    size='pop',
                    hover_name='country',
                    title="3D Plot of Life Expectancy, GDP per Capita, and Year (Gapminder Dataset)")

# Show the figure
fig.show()
```
