# Boxplot dari Data Random

untuk melihat distribusi data

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(100, 10, 200)

# Membuat boxplot
plt.boxplot(data)

# Menambahkan label sumbu y
plt.ylabel('Nilai')

# Menambahkan judul plot
plt.title('Contoh Boxplot')

# Menampilkan plot
plt.show()
```

```python
data = np.random.normal(100, 10, 200)

# Membuat boxplot
plt.boxplot(data, vert=False)

# Menambahkan judul plot
plt.title('Contoh Boxplot')

# Menampilkan plot
plt.show()
```

```python
# Data dummy untuk contoh
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)
data = [data1, data2, data3]

# Membuat boxplot
plt.boxplot(data)

# Menambahkan label sumbu x
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])

# Menambahkan label sumbu y
plt.ylabel('Nilai')

# Menambahkan judul plot
plt.title('Contoh Boxplot')

# Menampilkan plot
plt.show()
```

```python
data = [data1, data2, data3]

# Membuat boxplot
plt.boxplot(data, vert=False)

# Menambahkan label sumbu x
plt.yticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])

# Menambahkan label sumbu y
plt.xlabel('Nilai')

# Menambahkan judul plot
plt.title('Contoh Boxplot')

# Menampilkan plot
plt.show()
```

#Boxplot dari DataFrame

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/penjualan_toko.csv')

df
```

```python
df.columns
```

```python
# Mendapatkan total penjualan masing-masing bulan
kumpulan_bulan = df.columns[2:]

for bulan in kumpulan_bulan:
  df[bulan] = df[bulan]*df['Harga per Unit (Rp)']/1000000

df
```

```python
data = df[kumpulan_bulan]

plt.figure(figsize=(10,5))

plt.boxplot(data)

plt.title("Boxplot Penjualan Januari-Desember")
plt.ylabel("Dalam Juta Rupiah")
plt.xlabel("Bulan")

plt.show()
```
