1. Import data Pnejualan toko.csv ke dalam dataframe dengan nama **df**

```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/penjualan_toko.csv'
df = pd.read_csv(path)
df
```

Buatlah pie chart penjualan (harga per unit × bulan_target) berdasarkan bulan, gunakan variabel bulan_target dalam membuat pie chart, jika bulan_target diubah, maka pie chart juga akan berubah. Buatlah nilai penjualan terbesar menjadi explode atau keluar.

```python
bulan_target = 'Januari'
# Mengambil data penjualan
data = df['Harga per Unit (Rp)'] * df[bulan_target]
label = df['Jenis Bahan Bangunan']

param_exp = [0.1 if x == data.max() else 0 for x in data]

plt.figure(figsize=(6,6))
plt.title(f'Penjualan Bulan {bulan_target}')

plt.pie(data,
        labels=label,
        autopct="%.2f%%",
        pctdistance=0.8,
        explode=param_exp)

plt.show()
```
