# Mempersiapkan Data

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("tableau-colorblind10")

path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/shanghai_ranking_2024.csv'
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
filter = df['National/Regional Rank'] == '1'
df_top_universities = df[filter].head(15).sort_values('Alumni')
```

```python
df_top_universities['University_Name']
```

Kasus : membuat plot dari nama university dan alumni

```python
x = df_top_universities['University_Name']
y = df_top_universities['Alumni']

plt.bar(x, y)
```

```python
x = df_top_universities['University_Name']
y = df_top_universities['Alumni']

plt.title ('Nilai Alumni Shangai Rangking dari Universitas Terbaik Setiap Negara')
plt.xlabel ('Alumni')
plt.ylabel('Nama Universitas')

for index, value in enumerate(y):
  plt.text(value+1, index, str(value), color='b')

plt.barh(x, y)
```
