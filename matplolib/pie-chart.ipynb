```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/shanghai_ranking_2024.csv'
df = pd.read_csv(path)
df.head()
```

# Pie Chart Sederhana

```python
target = 'Harvard University'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:].values #mengambil data berdasarkan index
label = df[filter].iloc[0][3:].index
```

```python
plt.pie(data, labels = label)

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

# Modifikasi Tampilan Pie Chart

```python
target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:] #mengambil data berdasarkan index
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label)

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

Menambah Parameter Warna

```python
# atur warna
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:] #mengambil data berdasarkan index
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna)# setting warna

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

Memberikan nilai setiap persennya

```python
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna, autopct = '%.2f%%') #tambahkan autopct

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

```python
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna, autopct = '%.1f%%') #tambahkan autopct

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

Atur letak nilai presentasi

```python
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna, autopct = '%.1f%%',
        pctdistance = 0.8) #tambahkan pctdistance (0-1)

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

Mengubah lokasi label

```python
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna, autopct = '%.1f%%',
        pctdistance = 0.8,
        labeldistance = 0.5) #tambahkan label distance (0-1)

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

Memberikan efek bayangan

```python
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna, autopct = '%.1f%%',
        pctdistance = 0.8,
        labeldistance = 0.5,
        shadow = True) #tambahkan shadow

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```

Membuat pie terpisah

```python
warna = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'lightblue']
param_exp = [0, 0, 0.1, 0, 0, 0.2] # harus sesuai dengan jumlah pienya

target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels = label, colors = warna, autopct = '%.1f%%',
        pctdistance = 0.8,
        labeldistance = 0.5,
        shadow = True,
        explode = param_exp) #tambahkan parameter explode

plt.title(f"Nilai Indikator Shanghai Rangking \n{target}")

plt.show()
```
