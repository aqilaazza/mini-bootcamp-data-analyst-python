```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/fix.csv'
df = pd.read_csv(path, on_bad_lines='skip')
df.head()
```

```python
fig, ax = plt.subplots(1, 2)
```

```python
df.rename(columns={'Jumlah Anggaran': 'Jumlah Penduduk'}, inplace=True)
df.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 10), sharey=True)

ax[0].barh(df['Provinsi'], df['Realisasi Jumlah '] / 1000000)
ax[1].barh(df['Provinsi'], df['Realisasi Anggaran'] / 1000000000, color='g')

ax[0].set_title('Realisasi Jumlah KPM')
ax[1].set_title('Realisasi Anggaran KPM')

ax[0].set_xlabel('Realisasi Jumlah (Juta Orang)')
ax[1].set_xlabel('Realisasi Anggaran (Miliar Rupiah)')

fig.suptitle('DATA KPM TAHUN 2022', fontsize = 20)

fig.tight_layout()
fig.show()
```

```python
# Provinsi di Pulau Jawa
provinsi_jawa = [
    "DKI Jakarta",
    "Jawa Barat",
    "Banten",
    "Jawa Tengah",
    "DI Yogyakarta",
    "Jawa Timur"
]

# Provinsi di Pulau Kalimantan
provinsi_kalimantan = [
    "Kalimantan Barat",
    "Kalimantan Tengah",
    "Kalimantan Selatan",
    "Kalimantan Timur",
    "Kalimantan Utara"
]

# Provinsi di Pulau Sumatera
provinsi_sumatera = [
    "Aceh",
    "Sumatera Utara",
    "Sumatera Barat",
    "Riau",
    "Kepulauan Riau",
    "Jambi",
    "Bengkulu",
    "Sumatera Selatan",
    "Lampung",
    "Kepulauan Bangka Belitung"
]

# Provinsi di Pulau Sulawesi
provinsi_sulawesi = [
    "Sulawesi Utara",
    "Gorontalo",
    "Sulawesi Tengah",
    "Sulawesi Barat",
    "Sulawesi Selatan",
    "Sulawesi Tenggara"
]

fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2, figsize=(12, 8))

# Filter data untuk setiap pulau
df_jawa = df[df['Provinsi'].isin(provinsi_jawa)]
df_kalimantan = df[df['Provinsi'].isin(provinsi_kalimantan)]
df_sulawesi = df[df['Provinsi'].isin(provinsi_sulawesi)]
df_sumatera = df[df['Provinsi'].isin(provinsi_sumatera)]


ax1.barh(df_jawa['Provinsi'], df_jawa['Realisasi Anggaran']/1000000000, color='lightblue')
ax2.barh(df_kalimantan['Provinsi'], df_kalimantan['Realisasi Anggaran']/1000000000, color='lightgreen')
ax3.barh(df_sulawesi['Provinsi'], df_sulawesi['Realisasi Anggaran']/1000000000, color='pink')
ax4.barh(df_sumatera['Provinsi'], df_sumatera['Realisasi Anggaran']/1000000000, color='purple')

ax1.set_title('Provinsi di Pulau Jawa')
ax2.set_title('Provinsi di Pulau Kalimantan')
ax3.set_title('Provinsi di Pulau Sulawesi')
ax4.set_title('Provinsi di Pulau Sumatera')

fig.suptitle('KPM Tahun 2022', fontsize=24, x=0.58)
fig.supxlabel('Jumlah Realisasi Anggaran KPM (Miliyar Rupiah)')
fig.supylabel('Provinsi')

fig.tight_layout()

fig.show()
```

# Subplot2grid

subplot2grid adalah salah satu cara untuk membuat subplot (sub-figure) yang lebih fleksibel dalam Matplotlib. Fungsi subplot2grid memungkinkan Anda untuk menentukan grid dari subplot dan mengatur letak dari setiap subplot secara spesifik dalam grid tersebut

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 8

# Provinsi di Pulau Jawa
provinsi_jawa = [
    "DKI Jakarta",
    "Jawa Barat",
    "Banten",
    "Jawa Tengah",
    "DI Yogyakarta",
    "Jawa Timur"
]

# Provinsi di Pulau Kalimantan
provinsi_kalimantan = [
    "Kalimantan Barat",
    "Kalimantan Tengah",
    "Kalimantan Selatan",
    "Kalimantan Timur",
    "Kalimantan Utara"
]

# Provinsi di Pulau Sumatera
provinsi_sumatera = [
    "Aceh",
    "Sumatera Utara",
    "Sumatera Barat",
    "Riau",
    "Kepulauan Riau",
    "Jambi",
    "Bengkulu",
    "Sumatera Selatan",
    "Lampung",
    "Kepulauan Bangka Belitung"
]

# Provinsi di Pulau Sulawesi
provinsi_sulawesi = [
    "Sulawesi Utara",
    "Gorontalo",
    "Sulawesi Tengah",
    "Sulawesi Barat",
    "Sulawesi Selatan",
    "Sulawesi Tenggara"
]


path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/fix.csv'
df = pd.read_csv(path, on_bad_lines='skip')
df.head()
```

```python
df.rename(columns={'Jumlah Anggaran': 'Jumlah Penduduk'}, inplace=True)
df.head()
```

```python
df.rename(columns={'Realisasi Anggaran ': 'Realisasi Anggaran'}, inplace=True)
df.head()
```

```python
plt.figure(figsize=(15,8))

# Create a copy of the DataFrame to avoid SettingWithCopyWarning
df_plot = df.copy()

# Clean up column names from any leading/trailing whitespace
df_plot.columns = df_plot.columns.str.strip()

# Remove the 'Indonesia' row
df_plot = df_plot[df_plot['Provinsi'] != 'Indonesia']

ax1 = plt.subplot2grid((2,4), (0,0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid((2,4), (0,2))
ax3 = plt.subplot2grid((2,4), (0,3))
ax4 = plt.subplot2grid((2,4), (1,2))
ax5 = plt.subplot2grid((2,4), (1,3))

# Plotting with the cleaned column names
ax1.barh(df_plot['Provinsi'], df_plot['Jumlah Penduduk']/1000000, color='black', label='Jumlah Penduduk')
ax1.barh(df_plot['Provinsi'], df_plot['Realisasi Jumlah']/1000000, label='Jumlah Penerima KPM')
ax1.set_xlabel('Jumlah Orang (Dalam Juta)')
ax1.set_ylabel('Provinsi')
ax1.set_title('Perbandingan Jumlah Penduduk dan Penerima KPM')
ax1.legend()

# Filter data for each island from the corrected DataFrame
df_jawa = df_plot[df_plot['Provinsi'].isin(provinsi_jawa)]
df_kalimantan = df_plot[df_plot['Provinsi'].isin(provinsi_kalimantan)]
df_sumatera = df_plot[df_plot['Provinsi'].isin(provinsi_sumatera)]
df_sulawesi = df_plot[df_plot['Provinsi'].isin(provinsi_sulawesi)]

ax2.barh(df_jawa['Provinsi'], df_jawa['Realisasi Anggaran']/1000000000, color='lightblue')
ax3.barh(df_kalimantan['Provinsi'], df_kalimantan['Realisasi Anggaran']/1000000000, color='lightgreen')
ax4.barh(df_sumatera['Provinsi'], df_sumatera['Realisasi Anggaran']/1000000000, color='pink')
ax5.barh(df_sulawesi['Provinsi'], df_sulawesi['Realisasi Anggaran']/1000000000, color='orange')

ax2.set_xlabel('Realisasi Anggaran (Dalam Miliyar)')
ax3.set_xlabel('Realisasi Anggaran (Dalam Miliyar)')
ax4.set_xlabel('Realisasi Anggaran (Dalam Miliyar)')
ax5.set_xlabel('Realisasi Anggaran (Dalam Miliyar)')

ax2.set_title('Realisasi Anggaran \nPenerima KPM Pulau Jawa')
ax3.set_title('Realisasi Anggaran \nPenerima KPM Pulau Kalimantan')
ax4.set_title('Realisasi Anggaran \nPenerima KPM Pulau Sumatera')
ax5.set_title('Realisasi Anggaran \nPenerima KPM Pulau Sulawesi')

plt.suptitle("PENERIMA KPM 2022", fontsize=24)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

n_bins = 10
x = np.random.randn(1000, 3)

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

colors = ['red', 'tan', 'lime']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')

ax1.hist(x, n_bins, density=True, histtype='bar', stacked=True)
ax1.set_title('stacked bar')

ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('stack step (unfilled)')

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('different sample sizes')

fig.tight_layout()
plt.show()
```
