#Histogram Dari data Random

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(loc=50, scale=20, size=500)
```

```python
plt.hist(data, bins = 15, color='lightblue', edgecolor='black') #bins adalh jumlah data

plt.title('Contoh Histogram')
plt.xlabel('Nilai Data')
plt.ylabel('Frekuensi')

plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label='Mean') #ingin melihat rata rata
plt.axvline(np.median(data), color='green', linestyle='dashed', linewidth=2, label='Median') #ingin melihat median

plt.show()
```
