#Grid

Membuat plot yang kompleks dengan menyusun beberapa plot dalam bentuk grid. Grid plot sangat berguna untuk melakukan visualisasi komparatif atau ketika Anda ingin mengeksplorasi hubungan antara beberapa variabel secara bersamaan.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Menyiapkan data
taxis = sns.load_dataset("taxis")
```

```python
taxis.sample(4)
```

```python
taxis.describe(exclude='number')
```

#Facet Grids

FacetGrid adalah alat untuk menyusun beberapa plot dalam grid berdasarkan nilai satu atau lebih variabel kategorikal.

```python
sns.FacetGrid?
```

```python
sns.FacetGrid(taxis, row="color")
```

```python
g = sns.FacetGrid(taxis, col="color", row="payment")
g.map(sns.histplot, 'fare')
```

Taksi warna kuning lebih populer dari credit card maupun cash sedangakn hijau sedikit

```python
g = sns.FacetGrid(taxis, col="color", row="payment")
g.map(sns.scatterplot, 'fare', 'tip')
```

Taksi yang menggunakan payment cash tidak menggunakan tip, sedangkan yang menggunakan credit card ada yang tidak menggunakan tip namun kebanyak menggunkan tip

```python
g = sns.FacetGrid(taxis, col="pickup_borough")
g.map(sns.boxplot, 'total')

plt.tight_layout()
```

```python
g = sns.FacetGrid(taxis, col="pickup_borough")
g.map(sns.histplot, 'tip')

plt.tight_layout()
```

```python
g = sns.FacetGrid(taxis, col="pickup_borough")
g.map(sns.countplot, 'color')

plt.tight_layout()
```

# Pair Grid

PairGrid memungkinkan Anda membuat plot matriks yang menunjukkan hubungan pasangan-pasangan variabel dalam dataset.


```python
iris = sns.load_dataset("iris")
iris.head()
```

Screenshot 2025-07-28 113752.png

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris, hue='species')
g.map(sns.scatterplot)
g.add_legend()
```

```python
# Menggunakan dataset contoh dari seaborn
iris = sns.load_dataset("iris")

# Membuat PairGrid
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.histplot, kde=True)
```

#Join Grid

JointGrid memungkinkan Anda membuat plot bivariate yang menunjukkan hubungan antara dua variabel dengan plot marginals di sumbu x dan y. Ini adalah dasar dari jointplot.

```python
# Menggunakan dataset contoh dari seaborn
tips = sns.load_dataset("tips")

# Membuat JointGrid
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.kdeplot) #bisa gunakan histplot, kdeplot, distplot, dll
```

```python
# Menggunakan dataset contoh dari seaborn
tips = sns.load_dataset("tips")

# Membuat JointGrid
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
```
