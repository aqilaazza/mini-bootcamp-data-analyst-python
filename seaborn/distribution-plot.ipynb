#Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.load_dataset('tips')
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.histplot(data=tips, x='total_bill', bins = 30)
```

```python
sns.histplot(data=tips, x='total_bill', bins = 30, hue='time')
#hue mengatur 'berdasarkan'
```

```python
sns.histplot(data=tips, x='total_bill', bins = 30, hue='time', kde=True)

plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram Total Bill Berdasarkan Time')

plt.show()
```

#KDE

```python
plt.figure(figsize=(10, 5))

sns.kdeplot(data=tips, x='tip', hue='sex')

plt.xlabel('Tip (Dollar)')
plt.ylabel('Frequency')
plt.title('KDE Besar Tip')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.rugplot(data=tips, x='tip')

plt.xlabel('Tip (Dollar)')
plt.ylabel('Frequency')
plt.title('Rug Plot Besar Tip')

plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='time')
```

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips, hue='sex')
```
