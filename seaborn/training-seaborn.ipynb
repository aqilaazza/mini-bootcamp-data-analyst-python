1. 1. Buatlah dataframe dengan nama mpg dari dataset bawaan seaborn dengan nama yang sama!

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
mpg = sns.load_dataset('mpg')
```

2. Buatlah histogram dari kolom mpg dengan menyertakan garis kde!

```python
sns.histplot(data=mpg, x='mpg', kde=True)

plt.xlabel('MPG')
plt.ylabel('Total Jumlah')
plt.title('Histogram MPG')

plt.show()
```

3. Buatlah boxplot dari mpg berdasarkan tahun modelnya!

```python
sns.boxplot(x='model_year', y='mpg', data=mpg)

plt.xlabel('modal year')
plt.ylabel('mpg')
```

4. Buatlah pairplot dari kolom mpg, horsepower, dan acceleration berdasarkan originnya!

```python
sns.pairplot(data=mpg[['mpg','horsepower','acceleration', 'origin' ]], hue='origin', palette='viridis')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=mpg, x='model_year', y='mpg', hue='origin')

plt.title('Rata-rata MPG dari Tahun 1970-2982 \nBerdasarkan Negaranya')
plt.show()
```
