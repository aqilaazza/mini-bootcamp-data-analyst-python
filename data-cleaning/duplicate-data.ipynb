```python
import pandas as pd
```

```python
data = {
    'A' : [1, 2, 2, 1, 5],
    'B' : [5, 6, 6, 8, 9],
    'C' : [1, 2, 2, 1, 6]
}
df = pd.DataFrame(data)
df
```

Mengetahui Data Duplikat

```python
df.duplicated()
```

```python
filter = df.duplicated()
df[filter]
```

Mengatasi Data Duplikat

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset=['A', 'C']) #ingin menghapus duplikate berdasarkan kolom tertentu
```
