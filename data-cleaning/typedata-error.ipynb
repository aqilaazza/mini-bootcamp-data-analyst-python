```python
import pandas as pd

data = {
    'A': [1, 2, 3, 4],
    'B': ['5', '6', '7', '8'],
    'C': ['2020-01-01', '2020-02-01', '2020-03-01', '2020-04-01'],
    'D': ['1.1', '2.2', '3.3', '4.4']
}

df = pd.DataFrame(data)
df
```

Mendeteksi Kesalahan Tipe Data

```python
df.info()
```

Mengatasi Kesalahan Tipe Data

Menggunakan fungsi to_numerik dan to_datetime

```python
df['B'] = pd.to_numeric(df['B'])
df['D'] = pd.to_numeric(df['D'])
df['C'] = pd.to_datetime(df['C'])
```

```python
df.info()
```

Menggunakan Method .astype

```python
df = pd.DataFrame(data)
df.info()
```

```python
df['B'] = df['B'].astype('int64')
df['C'] = df['C'].astype('datetime64[ns]')
df['D'] = df['D'].astype('float64')
df.info()
```

Mengatasi data yang tidak bisa dikonversi

```python
data = {
    'A': [1, 2, 3, 4],
    'B': ['5', '6', '7', 'Hai']
}

df = pd.DataFrame(data)
df
```
