Fungsi info() pada Pandas digunakan untuk mendapatkan ringkasan singkat dari DataFrame, termasuk informasi tentang indeks, tipe data, dan jumlah nilai non-null untuk setiap kolom.

Datatype adalah tipe data yang ada dimiliki oleh setiap kolom dalam DataFrame:

int64: Bilangan bulat 64-bit. Contoh: 23, 45, 28

float64: Bilangan desimal 64-bit. Contoh: 23.5, 4.51, 24.0

object: Tipe data umum yang digunakan untuk teks atau campuran tipe. Contoh: 'A', 'B', 'C'

bool: Tipe data boolean yang hanya memiliki dua nilai: True atau False

datetime64: Tipe data untuk tanggal dan waktu. Contoh: '2024-01-01'

timedelta[ns]: Tipe data untuk durasi atau perbedaan waktu

category: Tipe data untuk data kategorikal yang dapat memiliki nilai tetap yang terbatas
