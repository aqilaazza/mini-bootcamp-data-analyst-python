```python
import pandas as pd
path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/Pokemon.csv'
df_pokemon = pd.read_csv(path, index_col='Num')
df_pokemon
```

Display Max Rows

untuk mengatur jumlah maksimal rows

```python
pd.options.display.max_rows = 10
df_pokemon
```

```python
pd.options.display.max_rows = 2
df_pokemon
```

```python
pd.options.display.max_rows = 2
print(df_pokemon)
```

Display Max Columns

```python
pd.options.display.max_columns = 1000
df_pokemon
```

```python
pd.options.display.max_columns = None
df_pokemon
```

Display Width

Mengatur lebar maksimal tampilan dataframe

```python
pd.options.display.width = 5
print(df_pokemon)
```

Display Precition

Mengatur jumlah angka desimal yang akan ditampilkan untuk angka floating point

```python
pd.options.display.max_rows = None
df_pokemon.mean(numeric_only=True)
```

```python
pd.options.display.max_rows = None
pd.options.display.precision = 2 #hanya digunakan 2 angka dibelakang koma
df_pokemon.mean(numeric_only=True)
```
