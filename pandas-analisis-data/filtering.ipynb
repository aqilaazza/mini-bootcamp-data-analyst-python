# Filtering DataFrame Part 1

Filtering
proses memilih baris data yang memenuhi kondisi tertentu

```python
import pandas as pd

path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/Pokemon.csv'
df = pd.read_csv(path, index_col = 'Num')
df.head(5)
```

```python
filter = df['Name'] == 'Pikachu'
df[filter]
```

```python
filter = df['Type 1'] == 'Fire'
df[filter]
```

# Filtering DataFrame Part 2

Filtering dengan operasi

```python
filter = df ['Attack'] == df['Attack'].max()
df[filter]
```

```python
filter = df ['Attack'] == df['Attack'].min()
df[filter]
```

```python
# mencari pokemon yang memiliki nilai attacks leboh dari nilai defense
filter = df['Attack'] > df['Defense']
df[filter]
```

```python
# mencari legendary yang bernilai true (boolean)
filter = df['Legendary'] == True
df[filter]
```

```python
# bertipe 1 dan total lebih dari 500
filter1 = df['Type 1'] == 'Fire'
filter2 = df['Total'] > 500
df[filter1 & filter2]
```

```python
# bertipe 1 atau total lebih dari 500
filter1 = df['Type 1'] == 'Fire'
filter2 = df['Total'] > 500
df[filter1 | filter2]
```
