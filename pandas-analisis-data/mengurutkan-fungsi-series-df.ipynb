```python
import pandas as pd
pd.options.display.max_columns = None

data = {
    'nama': ['Ansar', 'Usman', 'Hesti', 'Alex', 'Indra'],
    'gaji_dolar': [5300, 2300, 10500, 4000, 15000],
    'bonus': [100, 200, 300, 400, 500],
    'pengalaman': [1, 2, 3, 4, 5]
}

df = pd.DataFrame(data)
df
```

kasus : jika gajinya diatas 10000 akan dijadikan meneger

```python
def status(gaji):
  if gaji>10000:
    return 'manager'
  else:
    return 'staff'
```

```python
df['status'] = df['gaji_dolar'].apply(status)
df
```

kasus : jika dia staff akan difasilitasi motor dinas jika manager akan di fasilitasi mobil dinas

```python
def fasilitas(df):
  if df['status'] == 'manager':
    return 'mobil dinas'
  else:
    return 'motor dinas'

```

```python
df['fasilitas'] = df.apply(fasilitas, axis=1)
df
```

Kasus : Jika Manager pengalaman > 4 tahun maka akan dapat rumah dinas

```python
def fasilitas_lain(df):
  if df['status'] == 'manager':
    if df['pengalaman'] > 4:
      return 'rumah dinas'
    else:
      return 'tidak ada'
  else:
    return 'tidak ada'
```

```python
df['fasilitas_lain'] = df.apply(fasilitas_lain, axis=1)
df
```
