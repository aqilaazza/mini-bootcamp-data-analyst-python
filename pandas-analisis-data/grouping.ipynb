#Grouping DataFrame Part 1

```python
import pandas as pd

path = '/content/drive/MyDrive/Sertifikasi Data Analyst UDEMY/Sample Data/Pokemon.csv'
df = pd.read_csv(path, index_col = 'Num')
df.head(6)
```

```python
filter = df['Type 1'] == 'Grass'
df_filtered = df[filter]
df_filtered.head(4)
```

```python
df_filtered.mean(numeric_only=True)
```

```python
#proses gruping
#Gruping masing masing pokemon berdasarkan tipe satu (rata-rata)
grouped_df = df.groupby(by = 'Type 1')
grouped_df.mean(numeric_only=True)
```

```python
#menambahkan nilai attack pada masing masing tipe 1
grouped_df = df.groupby(by = 'Type 1')
grouped_df.mean(numeric_only=True) ['Attack']
```

```python
#gruping beberapa kolom
grouped_df = df.groupby(by = ['Type 1', 'Type 2'])
grouped_df.mean(numeric_only=True) ['Speed']
```

```python
grouped_df = df.groupby(by = ['Type 1', 'Type 2'])
grouped_df.count()
```

# Grouping DataFrame Part 2

Gruping dengan fungsi agregat

```python
# grouping dengan nilai rata rata total, nilai maximum attack dan nilai minimum defence
grouped_df = df.groupby(by = ['Type 1'])
grouped_df.agg({'Total' : 'mean', 'Attack' : 'max', 'Defense' : 'min'})
```

```python
#grup untuk nilai max dan min yang dimiliki tipe 1
grouped_df = df.groupby(by = ['Type 1'])
grouped_df.agg({'Total' : ['max', 'min']})
```

```python
#grup untuk nilai max, min total dan rata2 , standart attack yang dimiliki tipe 1 dan 2
grouped_df = df.groupby(by = ['Type 1', 'Type 2'])
grouped_df.agg({'Total' : ['max', 'min'], 'Attack' : ['mean', 'std']})
```

```python
# membuat fungsi sendiri contoh range dari masing maisng pokemon
def range(x):
  return x.max() - x.min()
```

```python
grouped_df = df.groupby(by = ['Type 1'])
grouped_df.agg({'Sp. Atk':range})
```
