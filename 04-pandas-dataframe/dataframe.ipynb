# Menghapus Baris dan Kolom DataFrame

```python
data_kendaraan = {
    'Merk': ['Toyota', 'Honda', 'Suzuki', 'Yamaha', 'Mitsubishi', 'Kawasaki', 'Daihatsu', 'Isuzu'],
    'Model': ['Avanza', 'Civic', 'Ertiga', 'NMAX', 'Pajero', 'Ninja', 'Xenia', 'Panther'],
    'Tahun Produksi': [2021, 2020, 2022, 2021, 2019, 2018, 2020, 2017],
    'Tipe': ['Mobil', 'Mobil', 'Mobil', 'Motor', 'Mobil', 'Motor', 'Mobil', 'Mobil']
}

mob_index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

df_kendaraan = pd.DataFrame(data_kendaraan, index = mob_index)

df_kendaraan['Warna'] = ['Merah', 'Biru', 'Silver', 'Hitam', 'Hitam', 'Merah', 'Putih', 'Putih']
df_kendaraan['Model Warna'] = df_kendaraan['Model'] + ' ' + df_kendaraan['Warna']

df_kendaraan
```

#Menggabungkan DataFrame

```python
import pandas as pd
```

Fungsi Concat

menggabungkan datafram berdasarkan kolom dan index

```python
# Membuat dua DataFrame
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['D', 'E', 'F'],
    'value1': [4, 5, 6]
})

df3 = pd.DataFrame({
    'value2': [4, 5, 6]
})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2]).reset_index(drop=True) # untuk merapikan index
```

```python
pd.concat([df1, df3], axis = 1) #1 artinya menggabungkan antar kolom
```

Fungsi Merge

menggabungkan table yang memiliki kolom yang sama

```python
dfA = pd.DataFrame({
    'key': ['A', 'B', 'C', 'D'],
    'value1': [1, 2, 3, 4]
})

dfB = pd.DataFrame({
    'key': ['A', 'B', 'E', 'F'],
    'value2': [5, 6, 7, 8]
})

dfC = pd.DataFrame({
    'value1': [1, 2, 3, 4]
}, index=['A', 'B', 'C', 'D'])

dfD = pd.DataFrame({
    'value2': [5, 6, 7, 8]
}, index=['A', 'B', 'E', 'F'])
```

```python
pd.merge(dfA, dfB, on='key')
#Nilai yang ada pada dfA dan dfB hanya A dan B saja maka yang akan di return hanya A dan B
```

```python
pd.merge(dfA, dfB, on='key', how='right') #untuk mengatur irisannya
```

Fungsi Join

Menggabungkan dataframe berdasarkan index

```python
dfC.join(dfD, how='left')
```

```python
dfC.join(dfD, how='inner')
```

```python
dfC.join(dfD, how='right')
```
